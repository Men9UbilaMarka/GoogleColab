Мы загружаем набор данных, который связан с расходом топᴫива и
выбросами угᴫекисᴫого газа автомобиᴫей. Затем мы раздеᴫяем наши данные на обучающие и
тестовые наборы, создаем модеᴫь с испоᴫьзованием обучающего набора, оцениваем вашу
модеᴫь с помощью тестового набора и, наконец, испоᴫьзуем модеᴫь дᴫя прогнозирования

1.  Подключаем нужные библитотеки

```python
import matplotlib.pyplot as plt
import pandas as pd
import pylab as pl
import numpy as np
%matplotlib inline
```

2. Загружаем датасет и отображаем его

```python
df = pd.read_csv("/content/FuelConsumption.csv")
df.head()
```

3. Обобщаем и упорядочиваем данные

```python
df.describe()
```

```python
cdf = df[['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_COMB','CO2EMISSIONS']]
cdf.head(9)
```

4. Визуализируем датасет

```python
viz = cdf[['CYLINDERS','ENGINESIZE','CO2EMISSIONS','FUELCONSUMPTION_COMB']]
viz.hist()
plt.show()
```

5. Строим график для каждой диаграммы с характерностью  к выбросам для определения линейности их связей

```python
plt.scatter(cdf.FUELCONSUMPTION_COMB, cdf.CO2EMISSIONS,color='blue')
plt.xlabel("FUELCONSUMPTION_COMB")
plt.ylabel("Emission")
plt.show()
```

```python
plt.scatter(cdf.ENGINESIZE, cdf.CO2EMISSIONS,color='blue')
plt.xlabel("Enginesize")
plt.ylabel("Emission")
plt.show()
```

**Практика**

1. Разделим датасет на обучающий и тестовый в соотношении 80% и 20%

```python
msk = np.random.rand(len(df)) < 0.8
train = cdf[msk]
test = cdf[~msk]
```

2. Визуализируем график зависимости выбросов

```python
plt.scatter(train.ENGINESIZE, train.CO2EMISSIONS,color='blue')
plt.xlabel("Enginesize")
plt.ylabel("Emission")
plt.show()
```

3. Испортируем необходимые библиотеки

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
train_x = np.asanyarray(train[['ENGINESIZE']])
train_y = np.asanyarray(train[['CO2EMISSIONS']])
regr.fit (train_x, train_y)

print ('Coefficients:', regr.coef_)
print ('Intercept:',regr.intercept_)
```

4. Строим линию соответвия по параметрам коэффициент и точка пересечения

```python
plt.scatter(train.ENGINESIZE, train.CO2EMISSIONS,color='blue')
plt.plot(train_x, regr.coef_[0][0]*train_x + regr.intercept_[0], '-r')
plt.xlabel("Enginesize")
plt.ylabel("Emission")
```

5. Вычислим точность модели

```python
from sklearn.metrics import r2_score
test_x = np.asanyarray(test[['ENGINESIZE']])
test_y = np.asanyarray(test[['CO2EMISSIONS']])
test_y_ = regr.predict(test_x)
print("Meanabsoluteerror:%.2f" % np.mean(np.absolute(test_y_ - test_y)))
print("Residualsumofsquares(MSE):%.2f" % np.mean((test_y_ - test_y) ** 2))
print("R2‑score:%.2f" % r2_score(test_y_ , test_y) )
```

**Вывод :** в данной прктической работе была использованна модель простой линейной регрессии (Simple Linear Regression). Это метод, который позволяет определить взаимосвязь между двумя переменными. Основная идея заключается в том, что одна переменная используется для прогнозирования другой переменной. Ее цель, найти линейную функцию, которая хорошо соответствует отношению между этими переменными. На примере данной практической мы на основании двух параметров: расход топлива и выбросы углекислого газа, обучили модель, а в его тестовой части мы с помощью линейной зависимости спрогнозировали определенный расход топлива при отпределенных выбросах углекислого газа. Преимущество этой модели заключается в математической простоте, но недостатков много: значение R2-score в 0.71 указывает на то, что модель объясняет большую часть вариации зависимой переменной (предыдущие модели имели лучший показатель), MAE и MSE в 22.66 и 879.89 указывают на приемлемую, но не идеальную точность модели.
